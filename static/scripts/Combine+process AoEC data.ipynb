```python
#lambda function to replace CID's country code with 3 letter ISO3
to3letters = lambda x: classifications_df.loc[x]['location_code']


#Extracts product data given a dataframe and product_id
def extract_product(df, i):
    temp=df;
    try:
        temp = temp.drop(['export_rca','cog','distance','hs_eci','hs_coi','sitc_eci','sitc_coi','pci'],axis=1)
    except:
        temp = temp.drop(['hs_eci','hs_coi','sitc_eci','sitc_coi','pci'],axis=1)
        
    temp = temp[temp['product_id']==i]
    temp = temp.drop('product_id',axis=1)
    
    try:
        temp['partner_id'] = temp['partner_id'].apply(to3letters)
    except:
        pass
    return temp;
```

```python
cph_df = pd.DataFrame()
for year in range(1995,2017):
    hs_filename = f"country_partner_hsproduct_year_{year}.csv"
    print(hs_filename)
    cph_df = cph_df.append(extract_product(pd.read_csv(hs_filename),753))
cph_df.describe()
```

```python
cph_df.to_csv('Soya beans trade HS 1995-2016.csv', encoding='utf-8', index=False)
print(len(cph_df))
cph_df.head()
```

```python
cps_df = pd.DataFrame()
for year in range(1962,2017):
    sitc_filename = f"country_partner_sitcproduct_year_{year}.csv"
    print(sitc_filename)
    cps_df = cps_df.append(extract_product(pd.read_csv(sitc_filename),763))
```
